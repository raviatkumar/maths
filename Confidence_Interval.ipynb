norm.ppf: This function is typically part of a Python library like SciPy or StatsModels and is used to calculate the percentile point of a standard normal (z) distribution. It stands for "percent point function," and it helps you find the z-score associated with a specific probability.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate sample data
np.random.seed(42)
population_mean = 100  # True population mean
population_std_dev = 15  # Known population standard deviation
sample_size = 30
sample_data = np.random.normal(population_mean, population_std_dev, size=sample_size)

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Calculate Z-score for a 95% confidence interval
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate margin of error
margin_of_error = z_critical * (population_std_dev / np.sqrt(sample_size))

# Calculate confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Plot the sample data distribution and confidence interval
plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=15, density=True, alpha=0.7, label='Sample Data')
plt.axvline(x=sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.axvline(x=confidence_interval[0], color='g', linestyle=':', label='Confidence Interval')
plt.axvline(x=confidence_interval[1], color='g', linestyle=':')
plt.title('Confidence Interval for Population Mean (Known Sigma)')
plt.xlabel('Sample Data')
plt.ylabel('Density')
plt.legend()
plt.show()

# Print results
print(f"Population Mean: {population_mean}")
print(f"Sample Mean: {sample_mean}")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

#one sample mean sigma is unkonw

The term "one-sample mean (σ unknown)" refers to a statistical scenario in which you are estimating the population mean of a single sample when the population standard deviation (σ) is unknown. This situation often arises when you have data from a single group or sample and you want to estimate the mean of the entire population from which the sample was drawn, but you don't know the population standard deviation.

To estimate the population mean in this scenario, you typically use the sample mean as your point estimate and the t-distribution to calculate confidence intervals.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for unbiased sample standard deviation
sample_size = len(sample_data)

# Calculate t-score for a 95% confidence interval
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)

# Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Calculate confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Plot the results
plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=10, alpha=0.7, color='blue', edgecolor='black', density=True, label='Sample Data')
plt.axvline(x=sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(x=confidence_interval[0], color='green', linestyle=':', label='Lower Bound')
plt.axvline(x=confidence_interval[1], color='green', linestyle=':', label='Upper Bound')
plt.title('One-Sample Mean (σ unknown) with Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

We calculate the margin of error using the sample standard deviation and square root of the sample size.

#One sample variance

Imagine you are conducting a survey to understand the age variability among voters in a local election. You randomly select 50 voters and record their ages. You want to estimate how much the ages of the voters vary from the average age.

Here's how you can calculate the one sample variance using this example:

Collect Sample Data: Survey 50 randomly selected voters and record their ages.

Calculate Sample Mean: Calculate the mean (average) of the ages of the 50 voters.

Calculate Deviations: For each voter's age, subtract the sample mean to calculate the deviation from the mean. This represents how much each voter's age differs from the average age.

Square Deviations: Square each deviation to get rid of negative values and emphasize larger deviations.

Calculate Sum of Squared Deviations: Sum up all the squared deviations.

Divide by Sample Size - 1: Divide the sum of squared deviations by (n - 1), where n is the sample size (50 in this case). This is the formula for calculating the sample variance.

```python
  import numpy as np
import matplotlib.pyplot as plt

# Simulate sample data (ages of 50 voters)
np.random.seed(42)
sample_ages = np.random.randint(18, 80, size=50)  # Simulated ages of voters

# Calculate sample mean
sample_mean = np.mean(sample_ages)

# Calculate squared deviations
squared_deviations = [(age - sample_mean)**2 for age in sample_ages]

# Calculate sum of squared deviations
sum_of_squared_deviations = sum(squared_deviations)

# Calculate sample variance (with Bessel's correction)
sample_size = len(sample_ages)
sample_variance = sum_of_squared_deviations / (sample_size - 1)

# Plot the data and sample variance
plt.figure(figsize=(10, 6))
plt.hist(sample_ages, bins=10, alpha=0.7, color='blue', edgecolor='black', density=True, label='Sample Ages')
plt.axvline(x=sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.title('Ages of Voters with Sample Variance')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()

print(f"Sample Variance: {sample_variance:.2f}")
```

Sample Variance: 343.67 indicates that the squared deviations of the ages of the 50 voters from their sample mean (average age) sum up to 343.67. This value provides a measure of the variability or spread of the ages in your sample.

A larger sample variance indicates that the ages of the voters in the sample are more spread out from the average age. In other words, there is a wider range of ages among the voters. On the other hand, a smaller sample variance suggests that the ages are relatively closer to the average age, indicating less variability.

In numerical terms, 343.67 represents the sum of the squared differences between each voter's age and the sample mean. However, it's important to note that the sample variance is given in squared units (e.g., years squared) because the deviations are squared. To get a sense of the variability in the same units as the original data, you would take the square root of the sample variance, which gives you the sample standard deviation.

Overall, the sample variance is a statistical measure that helps quantify the extent to which individual data points in a sample deviate from the sample mean, providing insights into the dispersion of the data.

In this example, the calculated sample variance gives you a measure of how much the ages of the voters in your sample vary from the average age. A larger sample variance indicates more variability among voter ages, while a smaller sample variance indicates that the ages are closer to the average. This information can be valuable for understanding the spread of ages among the voters and making informed decisions based on the characteristics of the sample.

#Two sample mean(σ1, σ2 unknown)

Example: Comparing the Income of Voters in Two Regions

Imagine you have data on the incomes of voters from two different regions: Region A and Region B. You want to determine if there is a significant difference in the mean incomes between the two regions.

Here's how you can perform a two-sample mean comparison with unknown population standard deviations using Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate sample data (incomes of voters in two regions)
np.random.seed(42)
sample_size = 150

income_region_a = np.random.normal(loc=60000, scale=10000, size=sample_size)  # Simulated incomes in Region A
income_region_b = np.random.normal(loc=55000, scale=12000, size=sample_size)  # Simulated incomes in Region B

# Calculate sample means
mean_income_a = np.mean(income_region_a)
mean_income_b = np.mean(income_region_b)

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(income_region_a, income_region_b)

# Plot the histograms of incomes for both regions
plt.figure(figsize=(10, 6))
plt.hist(income_region_a, bins=20, alpha=0.5, label='Region A')
plt.hist(income_region_b, bins=20, alpha=0.5, label='Region B')
plt.title('Incomes of Voters in Two Regions')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.legend()
plt.show()

print(f"Mean Income in Region A: {mean_income_a:.2f}")
print(f"Mean Income in Region B: {mean_income_b:.2f}")
print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")
```

In this code, we simulate the incomes of voters in two regions using normal distributions. We calculate the sample means for both regions and visualize the income distributions using histograms.

We then perform a two-sample t-test to assess whether there is a significant difference in mean incomes between Region A and Region B. The t-statistic and p-value from the t-test provide information about the significance of the observed difference in means.

The histograms show the income distributions of voters in both regions, and the t-test results help you determine whether the difference in mean incomes between the two regions is statistically significant. This analysis can provide insights into whether there is a meaningful income disparity between the two regions' voter populations.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate sample data (ages of voters in two regions)
np.random.seed(42)
sample_size = 200

age_region_a = np.random.normal(loc=45, scale=8, size=sample_size)  # Simulated ages in Region A
age_region_b = np.random.normal(loc=42, scale=7, size=sample_size)  # Simulated ages in Region B

# Calculate sample means
mean_age_a = np.mean(age_region_a)
mean_age_b = np.mean(age_region_b)

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(age_region_a, age_region_b)

# Plot the histograms of ages for both regions
plt.figure(figsize=(10, 6))
plt.hist(age_region_a, bins=20, alpha=0.5, label='Region A')
plt.hist(age_region_b, bins=20, alpha=0.5, label='Region B')
plt.title('Ages of Voters in Two Regions')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()

print(f"Mean Age in Region A: {mean_age_a:.2f}")
print(f"Mean Age in Region B: {mean_age_b:.2f}")
print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")
```

In this code, we simulate the ages of voters in two regions using normal distributions. We calculate the sample means for both regions and visualize the age distributions using histograms.

We then perform a two-sample t-test using the ttest_ind function from the scipy.stats module. The t-test assesses whether there is a significant difference in mean ages between Region A and Region B. The t-statistic and p-value from the t-test provide information about the significance of the observed difference in means.

The histograms show the age distributions of voters in both regions, and the t-test results help you determine whether the difference in mean ages between the two regions is statistically significant. This analysis can provide insights into whether there is a meaningful age difference between the two regions' voter populations.

Two sample means - Example
A motor company runs tests to investigate the fuel consumption of cars
using a newly developed fuel additive. Sixteen cars of the same make
and age are used, eight with the new additive and eight as controls. The
results, in miles per gallon over a test track under regulated conditions,
are as follows:


Control 27.0 32.2 30.4 28.0 26.5 25.5 29.6 27.2
Additive 31.4 29.9 33.2 34.4 32.0 28.7 26.1 30.3


Obtain a 95% confidence interval for the increase in miles per gallon
achieved by cars with the additive. State clearly any assumptions
required for this analysis.


To obtain a 95% confidence interval for the increase in miles per gallon achieved by cars with the additive, we can follow these steps:

Calculate the sample means and standard deviations for both the control and additive groups.

Calculate the difference in sample means.

Use the t-distribution to find the critical t-value for a 95% confidence level with 14 degrees of freedom (16 - 2).

Calculate the standard error of the difference in means.

Calculate the margin of error.

Construct the confidence interval.

Let's solve this problem using Python:

```python
import numpy as np
from scipy import stats

# Data for control and additive groups
control_data = np.array([27.0, 32.2, 30.4, 28.0, 26.5, 25.5, 29.6, 27.2])
additive_data = np.array([31.4, 29.9, 33.2, 34.4, 32.0, 28.7, 26.1, 30.3])

# Calculate sample means and standard deviations
mean_control = np.mean(control_data)
mean_additive = np.mean(additive_data)
std_control = np.std(control_data, ddof=1)
std_additive = np.std(additive_data, ddof=1)

# Calculate the difference in means
mean_difference = mean_additive - mean_control

# Calculate the t-critical value for a 95% confidence level
t_critical = stats.t.ppf(0.975, df=14)  # 14 degrees of freedom

# Calculate the standard error of the difference in means
std_error = np.sqrt((std_control**2 / 8) + (std_additive**2 / 8))

# Calculate the margin of error
margin_of_error = t_critical * std_error

# Construct the confidence interval
confidence_interval = (mean_difference - margin_of_error, mean_difference + margin_of_error)

# Print the results
print("Sample Mean (Control):", mean_control)
print("Sample Mean (Additive):", mean_additive)
print("Difference in Means:", mean_difference)
print("95% Confidence Interval:", confidence_interval)
```

In this code, we calculate the sample means and standard deviations for both the control and additive groups. We then calculate the difference in means and proceed to find the critical t-value for a 95% confidence level with 14 degrees of freedom (16 - 2).

The standard error of the difference in means is calculated using the pooled standard deviation formula. The margin of error is then calculated, and the confidence interval is constructed by adding and subtracting the margin of error from the mean difference.

The results provide the sample means, the difference in means, and the 95% confidence interval for the increase in miles per gallon achieved by cars with the additive.

#Two sample variance


Two-sample variance refers to comparing the variances of two independent samples from different populations. This helps determine whether there is a significant difference in the variability or spread of data between the two groups. Let's use an example related to voter ages to demonstrate two-sample variance visually using Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate sample data (ages of voters in two regions)
np.random.seed(42)
sample_size = 200

age_region_a = np.random.normal(loc=45, scale=8, size=sample_size)  # Simulated ages in Region A
age_region_b = np.random.normal(loc=45, scale=12, size=sample_size)  # Simulated ages in Region B

# Calculate sample variances
variance_a = np.var(age_region_a, ddof=1)
variance_b = np.var(age_region_b, ddof=1)

# Plot the ages distributions for both regions
plt.figure(figsize=(10, 6))
plt.hist(age_region_a, bins=20, alpha=0.5, label='Region A')
plt.hist(age_region_b, bins=20, alpha=0.5, label='Region B')
plt.title('Ages of Voters in Two Regions')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()

print(f"Sample Variance (Region A): {variance_a:.2f}")
print(f"Sample Variance (Region B): {variance_b:.2f}")
```

In this code, we simulate the ages of voters in two regions using normal distributions. We calculate the sample variances for both regions using the np.var function with the ddof=1 parameter for unbiased estimation.

We then visualize the age distributions of voters in both regions using histograms. The histograms show the spread of ages in each region.

The calculated sample variances provide information about the variability of voter ages in each region. Comparing the variances can help you assess whether there is a significant difference in the spread of ages between the two regions. If one region has a significantly larger variance than the other, it indicates greater variability in voter ages in that region.

#Two sample Variance

Two-sample proportion refers to comparing the proportions of a certain attribute or characteristic in two independent samples from different populations. This helps determine whether there is a significant difference in the proportion of the attribute between the two groups. Let's use an example related to voter preferences to explain two-sample proportion visually using Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate sample data (voter preferences in two regions)
np.random.seed(42)
sample_size = 200

preference_region_a = np.random.choice([0, 1], size=sample_size, p=[0.7, 0.3])  # Simulated preferences in Region A
preference_region_b = np.random.choice([0, 1], size=sample_size, p=[0.5, 0.5])  # Simulated preferences in Region B

# Calculate sample proportions
proportion_a = np.mean(preference_region_a)
proportion_b = np.mean(preference_region_b)

# Plot the voter preferences for both regions
plt.figure(figsize=(10, 6))
plt.hist(preference_region_a, bins=2, alpha=0.5, label='Region A')
plt.hist(preference_region_b, bins=2, alpha=0.5, label='Region B')
plt.title('Voter Preferences in Two Regions')
plt.xlabel('Preference (0 or 1)')
plt.ylabel('Frequency')
plt.legend()
plt.xticks([0, 1], ['Preference 0', 'Preference 1'])
plt.show()

print(f"Sample Proportion (Region A): {proportion_a:.2f}")
print(f"Sample Proportion (Region B): {proportion_b:.2f}")
```

In this code, we simulate voter preferences in two regions using random choices based on the given probabilities. The preferences are represented as binary values (0 or 1), where 0 indicates one preference and 1 indicates another preference.

We calculate the sample proportions of the preferred choice in each region by calculating the mean of the binary values.

The histograms visualize the distribution of voter preferences in both regions. The histograms show the proportions of the preferred choice (0 or 1) in each region.

The calculated sample proportions provide information about the relative proportions of voter preferences in each region. Comparing the proportions can help you assess whether there is a significant difference in the distribution of preferences between the two regions. If one region has a significantly different proportion of preferences compared to the other, it indicates a potential difference in voter preferences between the regions.
